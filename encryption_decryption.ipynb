```python
#encryption and decryption basics
#each letter is mapped to number/symbol
#a=,b=2,.....,z=26,space=*

#Encryption
def encrypt(text):
  result=[]
  for char in text.lower():
    if char==' ':                #space
      result.append('*')
    elif char.isupper():        #uppercase
      num=ord(char.lower())-ord('a')+1
      result.append('U'+str(num)) #mark U when uppercase it encountered
    elif char.islower():
      num=ord(char)-ord('a')+1
      result.append(str(num))
  return '-'.join(result)
#Decryption
def decrypt(code):
  result=""
  parts=code.split('-')
  for part in parts:
    if part=='*':
      result+=' '
    elif part.startswith('U'):
      num=int(part[1:])
      result+=chr(num+ord('a')-1).upper()
    else:
      num=int(part)
      result+=chr(num+ord('a')-1)
  return result

#demo
message=input("enter the message: ")
encryption=encrypt(message)
decryption=decrypt(encryption)
print("Original:",message)
print("Encrypted message:",encryption)
print("Decrypted message:",decryption)
```
